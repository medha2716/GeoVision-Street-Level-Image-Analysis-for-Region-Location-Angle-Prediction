```python

import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
from torchvision import transforms, models
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os
from tqdm import tqdm
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import LabelEncoder
import copy
import gc

# Constants
img_size_constant = (256, 256)
batch_size = 32
epochs = 100
patience = 15  # Increased patience for better convergence
lat_min, lat_max = 217800, 221700
lon_min, lon_max = 140000, 146500

# Device configuration
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

path = "/kaggle/input/smai-project-data/"
# path2 = "/kaggle/input/predicted-regions-updated/predicted_regions/"
path2 = "/kaggle/input/medha-kuch/predicted_regions_modded/"

# Load and preprocess data
train_df = pd.read_csv(path + "labels_train.csv")
val_df = pd.read_csv(path + "labels_val.csv")

predicted_region_train_df = pd.read_csv(path2+"predicted_regions_train.csv")
predicted_region_val_df = pd.read_csv(path2+"predicted_regions_val.csv")

# Merge with region predictions
train_df = pd.merge(train_df.drop('Region_ID', axis=1), 
                   predicted_region_train_df[['filename', 'Region_ID']], 
                   on='filename', how='inner')

val_df = pd.merge(val_df.drop('Region_ID', axis=1), 
                 predicted_region_val_df[['filename', 'Region_ID']], 
                 on='filename', how='inner')

# Filter out-of-bounds coordinates
train_df = train_df[(train_df['latitude'].between(lat_min, lat_max)) &
                    (train_df['longitude'].between(lon_min, lon_max))]

val_df = val_df[(val_df['latitude'].between(lat_min, lat_max)) &
                (val_df['longitude'].between(lon_min, lon_max))]

# Check data shapes
print(f"Training samples: {len(train_df)}")
print(f"Validation samples: {len(val_df)}")

# Encode regions
region_encoder = LabelEncoder()
train_df['region_encoded'] = region_encoder.fit_transform(train_df['Region_ID'])
val_df['region_encoded'] = region_encoder.transform(val_df['Region_ID'])

# Get number of unique regions
NUM_REGIONS = len(region_encoder.classes_)
print(f"Number of unique regions: {NUM_REGIONS}")

# Print class distribution
print("Region distribution in training set:")
region_counts = train_df['region_encoded'].value_counts()
print(region_counts)

# Normalization
lat_min, lat_max = train_df['latitude'].min(), train_df['latitude'].max()
lon_min, lon_max = train_df['longitude'].min(), train_df['longitude'].max()

for df in [train_df, val_df]:
    df['norm_latitude'] = (df['latitude'] - lat_min) / (lat_max - lat_min)
    df['norm_longitude'] = (df['longitude'] - lon_min) / (lon_max - lon_min)

print(f"Latitude range: {lat_min:.2f} to {lat_max:.2f}")
print(f"Longitude range: {lon_min:.2f} to {lon_max:.2f}")

# Enhanced Dataset Class with Region-Specific Augmentation
class GeoDataset(Dataset):
    def __init__(self, df, img_dir, is_train=False):
        self.df = df
        self.img_dir = img_dir
        self.is_train = is_train

        # Base transforms
        base_transforms = [
            transforms.Resize(img_size_constant),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ]

        # Train-time augmentations
        if self.is_train:
            # Create region statistics to adapt augmentation
            self.region_stats = {}
            for region in df['region_encoded'].unique():
                region_df = df[df['region_encoded'] == region]
                lat_std = region_df['latitude'].std()
                lon_std = region_df['longitude'].std()
                self.region_stats[region] = {
                    'lat_std': lat_std,
                    'lon_std': lon_std,
                    'count': len(region_df)
                }
            
            self.base_transform = transforms.Compose([
                transforms.RandomHorizontalFlip(p=0.3),
                transforms.RandomApply([
                    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.05)
                ], p=0.3),
                *base_transforms
            ])
            
            # Special transform for clahe
            self.clahe_transform = transforms.Compose([
                transforms.ToPILImage(),
                transforms.ToTensor(),
                transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
            ])
        else:
            # Validation transforms
            self.transform = transforms.Compose(base_transforms)

    def apply_clahe(self, img_tensor):
        """Apply CLAHE (Contrast Limited Adaptive Histogram Equalization) to improve contrast"""
        import cv2
        # Convert tensor to numpy array
        img = img_tensor.permute(1, 2, 0).numpy()
        # Convert to LAB color space
        lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
        # Apply CLAHE to L channel
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        lab[:,:,0] = clahe.apply(lab[:,:,0].astype(np.uint8))
        # Convert back to RGB
        enhanced = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)
        # Convert back to tensor
        return self.clahe_transform(enhanced)
    
    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.df.iloc[idx]['filename'])
        image = Image.open(img_path).convert('RGB')
        
        region = self.df.iloc[idx]['region_encoded']
        
        if self.is_train:
            image = self.base_transform(image)
            
            # Adaptively apply transforms based on region statistics
            region_info = self.region_stats[region]
            
            # Apply more aggressive augmentation to regions with more samples
            # and less augmentation to sparse regions
            if region_info['count'] > 100:  
                # For more populated regions, apply random rotation
                if np.random.random() < 0.4:
                    angle = np.random.uniform(-10, 10)
                    image = transforms.functional.rotate(image, angle)
                
                # Sometimes apply CLAHE to improve contrast in blurry images
                if np.random.random() < 0.3:
                    image = self.apply_clahe(image)
        else:
            image = self.transform(image)
    
        lat = torch.tensor(self.df.iloc[idx]['norm_latitude'], dtype=torch.float32)
        lon = torch.tensor(self.df.iloc[idx]['norm_longitude'], dtype=torch.float32)
        region_tensor = torch.tensor(region, dtype=torch.long)
    
        return image, region_tensor, torch.stack([lat, lon]), self.df.iloc[idx]['filename']

# Create datasets
train_dataset = GeoDataset(train_df, path + "images_train/images_train", is_train=True)
val_dataset = GeoDataset(val_df, path + "images_val/images_val", is_train=False)

# Calculate sampling weights for region balancing
region_counts = train_df['region_encoded'].value_counts()
weights = 1.0 / region_counts[train_df['region_encoded']].values
sampler = WeightedRandomSampler(weights, len(weights), replacement=True)

# Create data loaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=sampler)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

# Early stopping with model checkpointing
class EarlyStopper:
    def __init__(self, patience=1, min_delta=0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_score = float('inf')
        self.best_model = None
        
    def early_stop(self, val_score, model):
        if val_score < self.best_score:
            self.best_score = val_score
            self.counter = 0
            self.best_model = copy.deepcopy(model.state_dict())
            return False, True  # Not stopping, but model saved
        elif val_score > (self.best_score + self.min_delta):
            self.counter += 1
            if self.counter >= self.patience:
                return True, False  # Stopping, no model saved
            return False, False  # Not stopping, no model saved
        return False, False  # Not stopping, no model saved

# Improved model architecture with attention mechanism
class GeoModel(nn.Module):
    def __init__(self, num_regions, backbone_name='efficientnet_b1'):
        super(GeoModel, self).__init__()
        
        # Select backbone
        if backbone_name == 'efficientnet_b0':
            self.backbone = models.efficientnet_b0(pretrained=True)
            feature_dim = 1280
        elif backbone_name == 'efficientnet_b1':
            self.backbone = models.efficientnet_b1(pretrained=True)
            feature_dim = 1280
        elif backbone_name == 'efficientnet_b2':
            self.backbone = models.efficientnet_b2(pretrained=True)
            feature_dim = 1408
        elif backbone_name == 'resnet50':
            self.backbone = models.resnet50(pretrained=True)
            feature_dim = 2048
        else:
            raise ValueError(f"Unsupported backbone: {backbone_name}")
            
        self.backbone.classifier = nn.Identity()  # Remove classifier
        
        # Region embedding
        self.region_embedding = nn.Embedding(num_regions, 128)
        
        # Region-based attention mechanism
        self.attention = nn.Sequential(
            nn.Linear(128, 512),
            nn.ReLU(),
            nn.Linear(512, feature_dim),
            nn.Sigmoid()  # Output attention weights in [0,1]
        )
        
        # Feature fusion with attention
        self.fusion = nn.Sequential(
            nn.Linear(feature_dim + 128, 768),
            nn.BatchNorm1d(768),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(768, 384),
            nn.BatchNorm1d(384),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(384, 64),
            nn.ReLU(),
            nn.Dropout(0.1)
        )
        
        # Separate heads for latitude and longitude
        self.lat_head = nn.Linear(64, 1)
        self.lon_head = nn.Linear(64, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, x, regions):
        # Extract image features
        img_features = self.backbone(x)
        
        # Get region embeddings
        region_features = self.region_embedding(regions)
        
        # Generate attention weights based on region
        attention_weights = self.attention(region_features)
        
        # Apply attention to image features
        attended_features = img_features * attention_weights
        
        # Concatenate attended features with region embedding
        combined_features = torch.cat([attended_features, region_features], dim=1)
        
        # Shared feature processing
        shared_features = self.fusion(combined_features)
        
        # Separate predictions for latitude and longitude
        lat = self.sigmoid(self.lat_head(shared_features)).squeeze(-1)
        lon = self.sigmoid(self.lon_head(shared_features)).squeeze(-1)
        
        # Combine predictions
        return torch.stack([lat, lon], dim=1)

# Initialize model
model = GeoModel(NUM_REGIONS, backbone_name='efficientnet_b1').to(device)

# Custom loss function with weighted lat/lon components
class WeightedCoordinateLoss(nn.Module):
    def __init__(self, lat_weight=1.2, lon_weight=0.8):
        super(WeightedCoordinateLoss, self).__init__()
        self.lat_weight = lat_weight
        self.lon_weight = lon_weight
        self.mse = nn.MSELoss(reduction='none')
        
    def forward(self, outputs, targets):
        # Split predictions and targets into lat/lon components
        pred_lat, pred_lon = outputs[:, 0], outputs[:, 1]
        true_lat, true_lon = targets[:, 0], targets[:, 1]
        
        # Calculate separate losses
        lat_loss = self.mse(pred_lat, true_lat)
        lon_loss = self.mse(pred_lon, true_lon)
        
        # Apply weights
        weighted_loss = self.lat_weight * lat_loss + self.lon_weight * lon_loss
        
        return weighted_loss.mean()

# Loss and optimizer
criterion = WeightedCoordinateLoss(lat_weight=1.2, lon_weight=0.8)
optimizer = optim.Adam(model.parameters(), lr=1e-4, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', factor=0.5, patience=5, verbose=True)

# Early stopping
early_stopper = EarlyStopper(patience=patience)

# Track metrics
train_losses = []
val_losses = []
val_mse_history = []  # Stores (lat_mse, lon_mse, avg_mse)

# Training with mixed precision for faster training
from torch.cuda.amp import GradScaler, autocast

# Initialize gradient scaler for mixed precision training (if CUDA is available)
scaler = GradScaler() if device.type == 'cuda' else None

# Learning rate warmup
def get_lr_factor(epoch, warmup_epochs=5):
    if epoch < warmup_epochs:
        return (epoch + 1) / warmup_epochs
    return 1.0

# Region importance tracking
region_importance = torch.ones(NUM_REGIONS, device=device)

# Training loop with advanced techniques
print("Starting training...")
for epoch in range(epochs):
    # Adjust learning rate for warmup
    lr_factor = get_lr_factor(epoch)
    for param_group in optimizer.param_groups:
        param_group['lr'] = 1e-4 * lr_factor
    
    # Training phase
    model.train()
    running_loss = 0.0
    region_losses = torch.zeros(NUM_REGIONS, device=device)
    region_counts = torch.zeros(NUM_REGIONS, device=device)
    
    progress_bar = tqdm(train_loader, desc=f'Epoch {epoch+1}/{epochs}')
    
    for images, regions, targets, _ in progress_bar:
        images = images.to(device)
        regions = regions.to(device)
        targets = targets.to(device)
        
        optimizer.zero_grad()
        
        # Use mixed precision where supported
        if scaler is not None:
            with autocast():
                outputs = model(images, regions)
                loss = criterion(outputs, targets)
            
            # Scale loss and backprop
            scaler.scale(loss).backward()
            
            # Apply gradient clipping
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            
            # Update weights
            scaler.step(optimizer)
            scaler.update()
        else:
            # Regular training path (CPU or older GPUs)
            outputs = model(images, regions)
            loss = criterion(outputs, targets)
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            optimizer.step()
        
        # Track loss by region
        for i, r in enumerate(regions):
            region_losses[r] += criterion(outputs[i:i+1], targets[i:i+1]).item()
            region_counts[r] += 1
        
        running_loss += loss.item() * images.size(0)
        progress_bar.set_postfix({'loss': loss.item()})
    
    # Update region importance based on performance
    for r in range(NUM_REGIONS):
        if region_counts[r] > 0:
            avg_region_loss = region_losses[r] / region_counts[r]
            # Higher loss regions get higher importance (for weighted sampling)
            region_importance[r] = torch.clamp(avg_region_loss * 2.0, min=0.5, max=2.0)
    
    epoch_loss = running_loss / len(train_loader.dataset)
    train_losses.append(epoch_loss)
    
    # Print region performance stats every 10 epochs
    if epoch % 10 == 9:
        print("\nRegion performance:")
        for r in range(min(5, NUM_REGIONS)):  # Print first 5 regions
            if region_counts[r] > 0:
                print(f"Region {r}: Avg Loss = {region_losses[r]/region_counts[r]:.4f}, Importance = {region_importance[r]:.2f}, Samples = {region_counts[r]}")
        print("...")  # Indicate there are more regions
    
    # Validation phase
    if (epoch + 1) % 5 == 0 or epoch == epochs - 1:
        model.eval()
        val_loss = 0.0
        all_preds = []
        all_targets = []
        
        with torch.no_grad():
            for images, regions, targets, _ in val_loader:
                images = images.to(device)
                regions = regions.to(device)
                targets = targets.to(device)
                outputs = model(images, regions)
                
                val_loss += criterion(outputs, targets).item() * images.size(0)
                all_preds.append(outputs.cpu().numpy())
                all_targets.append(targets.cpu().numpy())
        
        val_loss /= len(val_loader.dataset)
        val_losses.append(val_loss)
        
        # Calculate MSEs
        all_preds = np.concatenate(all_preds)
        all_targets = np.concatenate(all_targets)
        
        pred_lats = all_preds[:, 0] * (lat_max - lat_min) + lat_min
        pred_lons = all_preds[:, 1] * (lon_max - lon_min) + lon_min
        true_lats = all_targets[:, 0] * (lat_max - lat_min) + lat_min
        true_lons = all_targets[:, 1] * (lon_max - lon_min) + lon_min
        
        lat_mse = mean_squared_error(true_lats, pred_lats)
        lon_mse = mean_squared_error(true_lons, pred_lons)
        avg_mse = 0.5 * (lat_mse + lon_mse)
        val_mse_history.append((lat_mse, lon_mse, avg_mse))
        
        print(f'\nValidation @ Epoch {epoch+1}:')
        print(f'Latitude MSE: {lat_mse:.2f}')
        print(f'Longitude MSE: {lon_mse:.2f}')
        print(f'Average MSE: {avg_mse:.2f}')
        print(f'Combined Val Loss: {val_loss:.4f}')
        
        # Update scheduler based on avg_mse
        scheduler.step(avg_mse)
        
        # Check early stopping
        stop, model_saved = early_stopper.early_stop(avg_mse, model)
        if model_saved:
            print(f"New best model saved with Avg MSE: {avg_mse:.2f}")
        if stop:
            print("Early stopping triggered!")
            break

# Load the best model
model.load_state_dict(early_stopper.best_model)
print(f"Loaded best model with Avg MSE: {early_stopper.best_score:.2f}")

# Clean up memory
gc.collect()
torch.cuda.empty_cache()

# Final evaluation
model.eval()
all_preds = []
all_targets = []
all_filenames = []

with torch.no_grad():
    for images, regions, targets, filenames in val_loader:
        images = images.to(device)
        regions = regions.to(device)
        outputs = model(images, regions)
        
        all_preds.append(outputs.cpu().numpy())
        all_targets.append(targets.cpu().numpy())
        all_filenames.extend(filenames)

all_preds = np.concatenate(all_preds)
all_targets = np.concatenate(all_targets)

pred_lats = all_preds[:, 0] * (lat_max - lat_min) + lat_min
pred_lons = all_preds[:, 1] * (lon_max - lon_min) + lon_min
true_lats = all_targets[:, 0] * (lat_max - lat_min) + lat_min
true_lons = all_targets[:, 1] * (lon_max - lon_min) + lon_min

lat_mse = mean_squared_error(true_lats, pred_lats)
lon_mse = mean_squared_error(true_lons, pred_lons)
avg_mse = 0.5 * (lat_mse + lon_mse)

print("\nFinal Validation Scores:")
print(f"Latitude MSE: {lat_mse:.2f}")
print(f"Longitude MSE: {lon_mse:.2f}")
print(f"Average MSE: {avg_mse:.2f}")

# Save the model
torch.save({
    'model_state_dict': model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
    'lat_min': lat_min,
    'lat_max': lat_max,
    'lon_min': lon_min,
    'lon_max': lon_max,
    'region_encoder': region_encoder,
    'avg_mse': avg_mse
}, 'best_geo_model.pth')

print("Best model saved to 'best_geo_model.pth'")

# Plot final results
plt.figure(figsize=(16, 12))

# Plot training and validation loss curves
plt.subplot(2, 2, 1)
plt.plot(train_losses, label='Training Loss')
plt.plot(np.arange(0, len(val_losses)) * 5, val_losses, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()

# Plot validation MSE
plt.subplot(2, 2, 2)
epochs = np.arange(1, len(val_mse_history)+1) * 5
plt.plot(epochs, [x[0] for x in val_mse_history], 'b-', label='Latitude MSE')
plt.plot(epochs, [x[1] for x in val_mse_history], 'g-', label='Longitude MSE')
plt.plot(epochs, [x[2] for x in val_mse_history], 'r-', label='Average MSE')
plt.xlabel('Epochs')
plt.ylabel('MSE')
plt.title('Validation MSE History')
plt.legend()

# Plot latitude predictions
plt.subplot(2, 2, 3)
plt.scatter(true_lats, pred_lats, alpha=0.5)
plt.plot([true_lats.min(), true_lats.max()], [true_lats.min(), true_lats.max()], 'k--')
plt.xlabel('Actual Latitude')
plt.ylabel('Predicted Latitude')
plt.title(f'Latitude Predictions (MSE: {lat_mse:.2f})')

# Plot longitude predictions
plt.subplot(2, 2, 4)
plt.scatter(true_lons, pred_lons, alpha=0.5)
plt.plot([true_lons.min(), true_lons.max()], [true_lons.min(), true_lons.max()], 'k--')
plt.xlabel('Actual Longitude')
plt.ylabel('Predicted Longitude')
plt.title(f'Longitude Predictions (MSE: {lon_mse:.2f})')

plt.suptitle(f'Final Model Performance (Average MSE: {avg_mse:.2f})', fontsize=16)
plt.tight_layout()
plt.savefig('geo_model_results.png')
plt.show()

# Create a prediction table for sample images
results_df = pd.DataFrame({
    'filename': all_filenames[:50],  # Show first 50 for brevity
    'true_lat': true_lats[:50],
    'pred_lat': pred_lats[:50],
    'true_lon': true_lons[:50],
    'pred_lon': pred_lons[:50],
    'lat_error': np.abs(true_lats[:50] - pred_lats[:50]),
    'lon_error': np.abs(true_lons[:50] - pred_lons[:50])
})

print("\nSample Predictions:")
print(results_df.head(10))
```
